```python
#Import pandas library

import pandas as pd
import seaborn as sb
sb.set_style("whitegrid")
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Read the dataset from csv file

data_all = pd.read_csv('Boston.csv')
```

```python
# Set up the dependent (labels) and the independent (features) variables
 
X_features = pd.DataFrame(data_all.iloc[:,:-1])
y_labels = pd.DataFrame(data_all.iloc[:,-1])
```

```python
# Have a glance at the independent variable

X_features
```

```python
# Have a glance at the dependent variable

y_labels
```

```python
# Split the data into train and test sets

from sklearn.model_selection import train_test_split

# 60% for training set, 40% for test set 
X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels, test_size= 0.4, random_state=1)


# Data for test set will be divided into two. 20% for test set and 20% for validation set
# X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=1)


# To print out the all data percentage. 
data = [y_train, y_test]
data_labels = ["Train: ", "Test: "]


for (data_label, dataset) in zip(data_labels,data):

    print ( (data_label) , round(len(dataset)/len(y_labels), 2))
```

```python
# in order to "fit" your model, you need to pass in train_features and train_labels.

# x_train.pop(x_train.columns[0]) # Remove Unnamed column
X_train.to_csv("train_features.csv", index=False)
y_train.to_csv("train_labels.csv", index=False)


# Once you have that stored model, you need to pass in test_features, val_features and test_labels and val_labels into predict, in order to generate new "predict"

# x_val.pop(x_val.columns[0])  # Remove Unnamed column
# x_val.to_csv("val_features.csv", index=False)
# y_val.to_csv("val_labels.csv", index=False)

# x_test.pop(x_test.columns[0])  # Remove Unnamed column
X_test.to_csv("test_features.csv", index=False)
y_test.to_csv("test_labels.csv", index=False)
```

```python
 # Have a glance at the shape of the train and test sets:
 
 # Shape of the train,  test and validation sets

print()
print ("X_train:", X_train.shape)
print ("y_train:", y_train.shape)
print ("X_test:", X_test.shape)
print ("y_test:", y_test.shape)
# print ("X_val:", X_val.shape)
# print ("y_val:", y_val.shape)
```

```python
# Train the algorithm

import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)  # # create object for the class

linear_regressor.fit(X_train, y_train) # perform linear regression

# X_train.head(10)
```

```python
# Having a look at the coefficients that the model has chosen:

v = pd.DataFrame(linear_regressor.coef_ , index = ["Coefficient"]).transpose()
w = pd.DataFrame(X_features.columns, columns=["Attribute"])
```

```python
# Concatenating the DataFrames to compare:

coeff_df = pd.concat([w, v], axis=1, join='inner')

coeff_df
```

```python
# Comparing the predicted value to the actual value:

# Predicted value

y_pred = linear_regressor.predict(X_test)
y_pred = pd.DataFrame(y_pred, columns=["Predicted"])

y_pred
```

```python
# Coefficient
"y = mx + b"
column_names1 = pd.DataFrame(X_test.iloc[0:0])

m = linear_regressor.coef_[0]

print ("Coefficients for each x values:")
for a,b in zip(column_names1, m):
    print(a , ": ", b)
```

```python
# Intercept

"y = mx + b"
column_names2 = pd.DataFrame(X_test.iloc[0:0])


b = linear_regressor.intercept_

print ("Intercept for each x values:")
for a,k in zip(column_names2, b):
    print(a , ": ", k)

```

```python
# following slope intercept form 

print("formula: y = {}x , {}".format(m, b) )


```

```python
# Evaluate algorithm

from sklearn import metrics
import numpy as np

# It is the difference between the measured value and “true” value
print ('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_pred))

# The Mean Squared Error measures how close a regression line is to a set of data points. 
print ('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_pred))

# RMSE is the standard deviation of the errors which occur when a prediction is made on a dataset
print ('Root Mean Squared Error: ', np.sqrt(metrics.mean_absolute_error(y_test, y_pred)))
```
